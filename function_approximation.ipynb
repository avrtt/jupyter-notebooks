### Approximate the given function using a polynomial of different degrees.
1. x1 = 1, x2 = 15
2. x1 = 1, x2 = 8, x3 = 15
3. x1 = 1, x2 = 4, x3 = 10, x4 = 15

```python
import numpy as np
import matplotlib.pyplot as plt
import math 

# function for approximation
def f(x):
    return np.sin(x / 5.) * np.exp(x / 10.) + 5. * np.exp(-x / 2.)

x = np.arange(1, 15, 0.1)
y = f(x)

plt.plot(x, y)
plt.show()
```

```python
# M*s = v

M1 = np.array([
    [1., 1.],
    [1., 15.]
])

v1 = np.array([f(1.), f(15.)])
print(v1)
```

```python
from scipy import linalg 

s1 = scipy.linalg.solve(M1, v1) 
print(s1) # the solution array
```

```python
print(s1[0] + s1[1]*1)
print(s1[0] + s1[1]*15)
```

```python
plt.plot(s1)
plt.show()
```

```python
M2 = np.array([
    [1., 1., math.pow(1., 2)],
    [1., 8., math.pow(8., 2)],
    [1., 15., math.pow(15., 2)]
])

v2 = np.array([f(1.), f(8.), f(15.)])
print(v2)
```

```python
s2 = scipy.linalg.solve(M2, v2)
print(s2)
```

```python
plt.plot(s2)
plt.show()
```

```python
M3 = np.array([
    [1., 1., math.pow(1., 2), math.pow(1., 3)],
    [1., 4., math.pow(4., 2), math.pow(4., 3)],
    [1., 10., math.pow(10., 2), math.pow(10., 3)],
    [1., 15., math.pow(15., 2), math.pow(15., 3)]
])

v3 = np.array([f(1.), f(4.), f(10.), f(15.)])
print(v3)
```

```python
s3 = scipy.linalg.solve(M3, v3)
print(s3)
```

```python
plt.plot(s3)
plt.show()
```
