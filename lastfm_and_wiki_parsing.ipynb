```python
import requests
from bs4 import BeautifulSoup
```

#### Target: all h2 and h3 HTML headings, [link](https://en.wikipedia.org/wiki/Bias%E2%80%93variance_tradeoff)

```python
r_1 = requests.get('https://en.wikipedia.org/wiki/Bias-variance_tradeoff')
soup_1 = BeautifulSoup(r_1.text, 'html.parser')
```

```python
for heads in soup_1.findAll('span', attrs={'class':'mw-headline'}):
    print(heads.text)
```

#### Target: titles of all articles in the category 'Machine Learning Algorithms', [link](https://en.wikipedia.org/wiki/Category:Machine_learning_algorithms)

```python
r_2 = requests.get('https://en.wikipedia.org/wiki/Category:Machine_learning_algorithms')
soup_2 = BeautifulSoup(r_2.text, 'html.parser')
```

```python
for i in soup_2.findAll('div', attrs={'class':'mw-category-group'}):
    for elem in i.findAll('li'):
        print(elem.text)
```

#### Target: names of all black metal bands from the list, [link](https://www.last.fm/tag/black+metal/artists)

```python
%%time

URLs = []
soup_3 = []

# getting all links
for page_num in range(1,49):
    URLs.append('https://www.last.fm/tag/black+metal/artists?page=' + str(page_num))

# extracting
for i in URLs:
    soup_3.append(BeautifulSoup((requests.get(i)).text, 'html.parser'))
    print('Page', i, 'is loaded.')
print('')
```

```python
# list of HTML texts
print(len(soup_3))
```

```python
# getting band names

bands = []

for p in range(0, 48):
    for heads in soup_3[p].findAll('h3'):
        bands.append(heads.text)
    bands.remove('Scrobble from Spotify?')
```

```python
for band in bands:
    print(band)
```
